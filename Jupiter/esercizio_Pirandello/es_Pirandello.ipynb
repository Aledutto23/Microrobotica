```python
import random

testo = open("sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt", "r")
  
d = dict()

for linea in testo:
    linea = linea.strip()
  
    linea = linea.lower()
  
    parole = linea.split(" ")
  
    for parola in parole:
        
        if parola in d:
            d[parola] += 1
            
        else:
            d[parola] = 1
        
            
for chiave in list(d.keys()):
    print(chiave, ":", d[chiave])
```
